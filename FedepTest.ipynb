```python
# read data
train_path = os.path.join('data', 'fmnist', 'data', 'train')   #femnist中200个客户端,synthetic_1_1中30个客户端
test_path = os.path.join('data', 'fmnist', 'data', 'test')
dataset = read_data(train_path, test_path)
```
